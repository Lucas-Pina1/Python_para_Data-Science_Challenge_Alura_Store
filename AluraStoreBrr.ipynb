### Importa칞칚o dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)


loja.head()

```

#1. An치lise do faturamento


```python
dados_lojas = {
    "Loja 1": loja,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}

# Calcula o faturamento total (Pre칞o * quantidade vendida, aqui assumimos apenas Pre칞o)
faturamento = {nome: df["Pre칞o"].sum() for nome, df in dados_lojas.items()}

# Formata para estilo brasileiro (ponto para milhar, v칤rgula para centavos)
faturamento_formatado = {
    loja: f"{valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    for loja, valor in faturamento.items()
}

# Exibe os faturamentos
for loja, valor in faturamento_formatado.items():
    print(f"{loja}: R$ {valor}")

plt.figure(figsize=(8, 5))
plt.bar(faturamento.keys(), faturamento.values(), color='skyblue')
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Lojas")
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import matplotlib.pyplot as plt

# Criar gr치fico de pizza para as categorias mais vendidas por loja
for nome_loja, df_loja in dados_lojas.items():
    # Conta quantas vezes cada categoria aparece (n칰mero de vendas por categoria)
    contagem_categorias = df_loja["Categoria do Produto"].value_counts()

    print(f"Vendas por Categoria em {nome_loja}:")
    print(contagem_categorias)

    # Criar gr치fico de pizza
    plt.figure(figsize=(7, 7))
    plt.pie(contagem_categorias, labels=contagem_categorias.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"游꼣 Distribui칞칚o das Categorias Vendidas - {nome_loja}")
    plt.axis('equal')  # Deixa o gr치fico em formato de c칤rculo
    plt.tight_layout()
    plt.show()
```

# 3. M칠dia de Avalia칞칚o das Lojas

```python
# Dicion치rio com a m칠dia das avalia칞칫es por loja
avaliacoes = {nome: df["Avalia칞칚o da compra"].mean() for nome, df in dados_lojas.items()}

print("M칠dia de Avalia칞칚o por Loja:")
for loja, media in avaliacoes.items():
    print(f"{loja}: {media:.2f}")

# Criando gr치fico de linha com melhorias visuais
plt.figure(figsize=(8, 5))
plt.plot(
    list(avaliacoes.keys()),
    list(avaliacoes.values()),
    marker='o',
    color='green',
    linewidth=2,
    linestyle='-',
    label='Nota m칠dia'
)

# Adiciona r칩tulos em cada ponto
for loja, nota in avaliacoes.items():
    plt.text(loja, nota + 0.05, f"{nota:.2f}", ha='center', fontsize=10)

# Ajustes visuais
plt.title("救 Avalia칞칚o M칠dia por Loja", fontsize=14)
plt.ylabel("Nota m칠dia (0 a 5)")
plt.xlabel("Lojas")
plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Criar gr치ficos de barras para cada loja
for nome_loja, df_loja in dados_lojas.items():
    # Contagem de produtos
    contagem_produtos = df_loja["Produto"].value_counts()

    # Top 5 mais vendidos
    top_mais_vendidos = contagem_produtos.head(5)

    # Top 5 menos vendidos (inverte e pega os 칰ltimos 5)
    top_menos_vendidos = contagem_produtos.tail(5)

    print(f"Top 5 Produtos Mais Vendidos em {nome_loja}:")
    print(top_mais_vendidos)
    print("\nTop 5 Produtos Menos Vendidos em", nome_loja)
    print(top_menos_vendidos)

    # Gr치fico - Mais vendidos
    plt.figure(figsize=(8, 4))
    plt.bar(top_mais_vendidos.index, top_mais_vendidos.values, color='seagreen')
    plt.title(f"游댛 Top 5 Produtos Mais Vendidos - {nome_loja}")
    plt.xlabel("Produto")
    plt.ylabel("Vendas")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Gr치fico - Menos vendidos
    plt.figure(figsize=(8, 4))
    plt.bar(top_menos_vendidos.index, top_menos_vendidos.values, color='salmon')
    plt.title(f"游댵 Top 5 Produtos Menos Vendidos - {nome_loja}")
    plt.xlabel("Produto")
    plt.ylabel("Vendas")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

# 5. Frete M칠dio por Loja


```python
frete_medio = {nome: df["Frete"].mean() for nome, df in dados_lojas.items()}

print("Frete M칠dio por Loja:")
for loja, frete in frete_medio.items():
    print(f"{loja}: R$ {frete:.2f}")

# Criar gr치fico de barras
plt.figure(figsize=(8, 5))
plt.bar(frete_medio.keys(), frete_medio.values(), color='orange')

# Adicionar os valores no topo das barras
for loja, valor in frete_medio.items():
    plt.text(loja, valor + 1, f"R${valor:.2f}", ha='center', fontsize=10)

# Configura칞칫es visuais
plt.title("游뚴 Frete M칠dio por Loja", fontsize=14)
plt.xlabel("Lojas")
plt.ylabel("Valor m칠dio do frete (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
